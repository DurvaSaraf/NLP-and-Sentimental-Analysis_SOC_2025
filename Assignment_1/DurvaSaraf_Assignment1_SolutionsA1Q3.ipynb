```python
df = pd.read_csv('Messages.csv', header=None, names=['Messages', 'Spam/Ham']) # no header so added


print(df.head()) # confirmimng its fixed

```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer


from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train,X_test,y_train,y_test = train_test_split(df['Messages'],df['Spam/Ham'],test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
type(X_train)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the vectorizer
vectorizer = TfidfVectorizer()

                                                # Fit on training data and transform both train and test
X_train_vec = vectorizer.fit_transform(X_train) # they are now nemerical matrices
X_test_vec = vectorizer.transform(X_test)
```

```python
      print(X_train_vec)
```

```python
print(X_test_vec)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

# Train Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train_vec, y_train)

# Train Naive Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# Predictions
lr_pred = lr_model.predict(X_test_vec)
nb_pred = nb_model.predict(X_test_vec)

# Evaluation
print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_test, lr_pred))
print("F1 Score:", f1_score(y_test, lr_pred, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_test, lr_pred))

print("\nNaive Bayes:")
print("Accuracy:", accuracy_score(y_test, nb_pred))
print("F1 Score:", f1_score(y_test, nb_pred, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_test, nb_pred))
```
